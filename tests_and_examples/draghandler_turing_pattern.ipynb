```python
%matplotlib qt
import matplotlib.animation as animation
from matplotlib.widgets import Slider, Button
import matplotlib as mpl
from matplotlib import pyplot as plt
import scipy.interpolate as inter
import numpy as np
import scipy.stats as stats
import torch

size = 200
r = 20

def laplacian(Z):
    return (
        -4 * Z
        + np.roll(Z, (0, -1), (0, 1))
        + np.roll(Z, (0, 1), (0, 1))
        + np.roll(Z, (-1, 0), (0, 1))
        + np.roll(Z, (1, 0), (0, 1))
    )

# function for getting new y data from updated params
def acquire_ydata(params):
   global U
   global V
   for _ in range(1000):
      Lu = laplacian(U)
      Lv = laplacian(V)
      reaction = U * V * V
      Du = params[0]
      Dv = params[1]
      F = params[2]
      k = params[3]

      U += Du * Lu - reaction + F * (1 - U)
      V += Dv * Lv + reaction - (F + k) * V


def init_ydata(params):
   global U
   global V
   init_u= params[4]
   init_v = params[5]
   U = np.ones((size, size))
   V = np.zeros((size, size))

   U[size//2 - r:size//2 + r, size//2 - r:size//2 + r] = init_u
   V[size//2 - r:size//2 + r, size//2 - r:size//2 + r] = init_v
   
# the values for the initial plot
params_names = ['diff_rate_u', 'diff_rate_v', 'feed', 'kill', 'init_u', 'init_v']
# same sequence as the params_names
params = [0.16, 0.08, 0.060, 0.062, 0.5, 0.25]


# plot the initial params
init_ydata(params)


fig = plt.figure(figsize=(12, 6))
gs = fig.add_gridspec(1, 2, width_ratios=[3, 1])
ax_img = fig.add_subplot(gs[0])
m1 = ax_img.imshow(V, cmap='plasma')
ax_img.set_title("Turing Pattern")
ax_img.axis('off')


def update(val):
    # update curve
    global params
    global U
    global V

    for i in range(len(params)):
      params[i] = sliders[i].val 

def generate_pattern(event):
    global params
    global U
    global V
    init_ydata(params)
    acquire_ydata(params)
    m1.set_data(V)  # Update image data with the new pattern
    fig.canvas.draw_idle()



def reset(event):
    global params
    global U
    global V
    
    #reset the values
    for i in np.arange(len(params)):
      sliders[i].reset()
    init_ydata(params)
    # acquire_ydata(params)
    m1.set_data(V) 
    # redraw canvas while idle
    fig.canvas.draw_idle()




sliders = []
for i in range(len(params)):
    axamp = plt.axes([0.68, 0.8 - i * 0.1, 0.18, 0.1])
    s = Slider(axamp, params_names[i], 0, 3, valinit=params[i])
    sliders.append(s)
    s.on_changed(update)

# Button for generating pattern
axgen = plt.axes([0.5, 0.06, 0.12, 0.03])
gen_button = Button(axgen, 'Generate Pattern')
gen_button.on_clicked(generate_pattern)

# Reset Button
axres = plt.axes([0.5, 0.01, 0.12, 0.03])
reset_button = Button(axres, 'Reset')
reset_button.on_clicked(reset)

plt.show()
```
