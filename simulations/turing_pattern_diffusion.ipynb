```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Mon Mai 19

@author: ChatGPT
"""

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Grid size
size = 200
# Diffusion rates
Du, Dv = 0.16, 0.08
# Feed and kill rates
F, k = 0.060, 0.062

# Initialize concentrations
U = np.ones((size, size))
V = np.zeros((size, size))

# Seed initial V chemical
r = 20
U[size//2 - r:size//2 + r, size//2 - r:size//2 + r] = 0.50
V[size//2 - r:size//2 + r, size//2 - r:size//2 + r] = 0.25

def laplacian(Z):
    return (
        -4 * Z
        + np.roll(Z, (0, -1), (0, 1))
        + np.roll(Z, (0, 1), (0, 1))
        + np.roll(Z, (-1, 0), (0, 1))
        + np.roll(Z, (1, 0), (0, 1))
    )

# Run simulation
for i in range(1000):
    Lu = laplacian(U)
    Lv = laplacian(V)
    reaction = U * V * V
    U += Du * Lu - reaction + F * (1 - U)
    V += Dv * Lv + reaction - (F + k) * V
 

plt.imshow(V, cmap='plasma')
plt.axis('off')
plt.title(f'Step {i}')
plt.show()

```
